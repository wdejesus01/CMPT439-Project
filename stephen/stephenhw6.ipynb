```python
def jacobi(matrice,tolval,stopcriterion,n):
    
    
    coefficent = matrice[:,:-1] # coeffiecent
    augmented = matrice[:,-1] #augmented
    
    noveaucoe = coefficent.copy()#copying coefficent
    noveauaug = augmented.copy()#copying augmented 
    for i in range(n):#checking if val is diagonally dominant or not
        offsum = 0#retreiving sum
        diagnonal = np.abs(noveaucoe[i,i])#
        for j in range(n):
            if(i!=j):#if it is not, just get sum of coefficents
                offsum +=np.abs(noveaucoe[i,j])
        if diagnonal < offsum:
            print("not diagonally dominant. Must transform")#if the diagonal is smaller than sum,display this message
            
        
    for i in range(n):
        noveauaug[i]=noveauaug[i]/noveaucoe[i,i]#transforming biasses via dividng them by diagonal element
        for j in range(n):
            if i !=j:# avoiding redundant operations in main loop
                noveaucoe[i,j]= noveaucoe[i,j]/noveaucoe[i,i]
    newval=np.ones(n)
    oldval=np.ones(n)        
    error=10#starting new while loop
    while(error>tolval):#main loop for our iterative process
        error = 0#initalizer of error
        for i in range(n):
            oldval[i]=newval[i]#store current approximations
            newval[i] = noveauaug[i]#storing current approximations
        for i in range(n):#loop for calculations
            for j in range(n):
                if(i!=j):
                    newval[i]=newval[i]-noveaucoe[i,j]*oldval[j]
                    
        if stopcriterion == "MAE":#if criterion is MAE
            error = np.mean(np.abs(newval-oldval))
        elif stopcriterion == "RMSE":# if critieion is RMSE
            error = np.sqrt(np.mean(np.abs(newval-oldval)**2))
        else:
            return None
    return newval
```

```python
def gauss(matrice,tolval,stopcriterion,n):
    coefficent = matrice[:,:-1] # coeffiecent
    augmented = matrice[:,-1] #augmented
    
    noveaucoe = coefficent.copy()
    noveauaug = augmented.copy()
    #checking if dagonaly or not
    for i in range(n):
        offsum = 0
        diagnonal = np.abs(noveaucoe[i,i])
        for j in range(n):
            if(i!=j):
                offsum +=np.abs(noveaucoe[i,j])
        if diagnonal < offsum:
            print("not diagonally dominant. Must transform")
            
    for i in range(n):
        noveauaug[i]=noveauaug[i]/noveaucoe[i,i]
        for j in range(n):
            if(i!=j):
                 noveaucoe[i,j]= noveaucoe[i,j]/ noveaucoe[i,i]
    x=np.ones(n)
    oldval=np.ones(n)
    error = 10#error intializer is assigned to 10
    
    while(error > tolval):
        error=0# initalzatior of error
        oldval=x.copy()
        for i in range(n):
            x[i]=noveauaug[i]#updating current approximations of new x array
            for j in range(n):
                if i!= j:
                    x[i]=x[i]- noveaucoe[i,j]*x[j]#latest avaliable approximations
                    
            #stop criterions
        if stopcriterion == "MAE":
            error = np.mean(np.abs(x-oldval))
        elif stopcriterion == "RMSE":
            error = np.sqrt(np.mean(np.abs(x-oldval)**2))
        else:
            return None   
        
        
    return x
```

```python
import numpy as np

def is_diagonally_dominant(A):
    """
    Checks if a square matrix A is weakly diagonally dominant.
    A(n,n) is diagonally dominant if for all i=1..n,
    |A[i,i]| >= sum(|A[i,j]|) for j != i
    """
    n = A.shape[0]
    for i in range(n):
        diag_element = abs(A[i, i])
        off_diag_sum = 0
        for j in range(n):
            if i != j:
                off_diag_sum += abs(A[i, j])
        
        if diag_element < off_diag_sum:
            return False
    return True

def jacobi(aug_matrix, tol, stop_criterion, max_iter=100):
    """
    Solves a system of linear equations Ax = b using the Jacobi iterative method.

    Args:
        aug_matrix (np.array): The augmented matrix [A|b].
        tol (float): The tolerance threshold for the stopping criterion.
        stop_criterion (str): 'mae' for Mean Absolute Error or 'rmse' for 
                              Root Mean Square Error.
        max_iter (int, optional): Maximum number of iterations. Defaults to 100.

    Returns:
        tuple: (solution_vector, iterations_count)
               Returns (None, 0) if the method cannot proceed (e.g., 0 on diagonal).
    """
    A = aug_matrix[:, :-1]
    b = aug_matrix[:, -1]
    n = len(b)

    # Check for zeros on the diagonal
    if np.any(np.diag(A) == 0):
        print("Error (Jacobi): A diagonal element is zero. Method cannot proceed.")
        return None, 0

    # Check for diagonal dominance
    if not is_diagonally_dominant(A):
        print("\n---------------------------------------------------------")
        print("Warning (Jacobi): The coefficient matrix is not diagonally dominant.")
        print("Convergence is not guaranteed. Proceeding anyway.")
        print("---------------------------------------------------------\n")

    # Initialize solution vector x with ones
    x = np.ones(n)
    x_new = np.zeros(n)
    
    iter_count = 0
    error = float('inf')

    while error > tol and iter_count < max_iter:
        iter_count += 1
        
        # Jacobi iteration formula
        for i in range(n):
            sigma = 0
            for j in range(n):
                if i != j:
                    sigma += A[i, j] * x[j]  # Use values from *previous* iteration (x)
            x_new[i] = (b[i] - sigma) / A[i, i]

        # Calculate error
        diff = x_new - x
        if stop_criterion == 'mae':
            error = np.mean(np.abs(diff))
        elif stop_criterion == 'rmse':
            error = np.sqrt(np.mean(diff**2))
        else:
            raise ValueError("Invalid stop_criterion. Use 'mae' or 'rmse'.")

        # Update x for the next iteration
        x = x_new.copy()

    if iter_count == max_iter and error > tol:
        print(f"Warning (Jacobi): Method failed to converge within {max_iter} iterations.")
        
    return x, iter_count

def gauss_seidel(aug_matrix, tol, stop_criterion, max_iter=100):
    """
    Solves a system of linear equations Ax = b using the Gauss-Seidel iterative method.

    Args:
        aug_matrix (np.array): The augmented matrix [A|b].
        tol (float): The tolerance threshold for the stopping criterion.
        stop_criterion (str): 'mae' for Mean Absolute Error or 'rmse' for 
                              Root Mean Square Error.
        max_iter (int, optional): Maximum number of iterations. Defaults to 100.

    Returns:
        tuple: (solution_vector, iterations_count)
               Returns (None, 0) if the method cannot proceed (e.g., 0 on diagonal).
    """
    A = aug_matrix[:, :-1]
    b = aug_matrix[:, -1]
    n = len(b)

    # Check for zeros on the diagonal
    if np.any(np.diag(A) == 0):
        print("Error (Gauss-Seidel): A diagonal element is zero. Method cannot proceed.")
        return None, 0

    # Check for diagonal dominance
    if not is_diagonally_dominant(A):
        print("\n---------------------------------------------------------")
        print("Warning (Gauss-Seidel): The coefficient matrix is not diagonally dominant.")
        print("Convergence is not guaranteed. Proceeding anyway.")
        print("---------------------------------------------------------\n")

    # Initialize solution vector x with ones
    x = np.ones(n)
    
    iter_count = 0
    error = float('inf')

    while error > tol and iter_count < max_iter:
        iter_count += 1
        x_old = x.copy() # Store old x for error calculation
        
        # Gauss-Seidel iteration formula
        for i in range(n):
            sigma = 0
            for j in range(n):
                if i != j:
                    # Uses *new* x[j] for j < i and *old* x[j] for j > i
                    sigma += A[i, j] * x[j] 
            x[i] = (b[i] - sigma) / A[i, i]

        # Calculate error
        diff = x - x_old
        if stop_criterion == 'mae':
            error = np.mean(np.abs(diff))
        elif stop_criterion == 'rmse':
            error = np.sqrt(np.mean(diff**2))
        else:
            raise ValueError("Invalid stop_criterion. Use 'mae' or 'rmse'.")

    if iter_count == max_iter and error > tol:
        print(f"Warning (Gauss-Seidel): Method failed to converge within {max_iter} iterations.")
        
    return x, iter_count
```
